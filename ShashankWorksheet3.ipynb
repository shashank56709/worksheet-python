```python
import pandas as pd

data_file='/content/drive/MyDrive/Python Sem 3 Workshop /dataset/Titanic-Dataset.csv'
df = pd.read_csv(data_file)

#1
fare = df[['Fare']]
print('Dataframe of Fare')
print(fare.head())

class_age = df[['Pclass', 'Age']]
print("Class and Age DataFrame:")
print(class_age.head())

survived_gender = df[['Survived', 'Sex']]
print("\nSurvived and Gender DataFrame:")
print(survived_gender.head())

#2
fare_gt_100 = df[df['Fare'] > 100]
print("Passengers with fare greater than 100:")
print(fare_gt_100.head())

first_class = df[df['Pclass'] == 1]
print("Passengers in first class:")
print(first_class.head())

female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]
print("\nFemale passengers under 18:")
print(female_under_18.head())

#3.1
df['Fare per Year'] = df['Fare'] / df['Age']
high_fare_age = df[df['Fare per Year'] > 5]
high_fare_age_sort = high_fare_age.sort_values(by='Fare per Year', ascending=False)
result = high_fare_age_sort[['Name', 'Fare per Year']]
print("Result for high fare per year:")
print(result.head())

df['Fare per Class'] = df['Fare'] / df['Pclass']
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]
adult_males_sort = adult_males.sort_values(by='Fare per Class', ascending=False)
result = adult_males_sort[['Name', 'Age', 'Fare per Class']]
print("Result for adult male passengers sorted by fare per class:")
print(result.head())

#3.2
total_fare = df['Fare'].sum()
print(total_fare)
first_class_fare = df[df['Pclass'] == 1]['Fare'].sum()
print(first_class_fare)
second_class_fare = df[df['Pclass'] == 2]['Fare'].sum()
print(second_class_fare)
third_class_fare = df[df['Pclass'] == 3]['Fare'].sum()
print(third_class_fare)
fare_totals_by_class = [first_class_fare, second_class_fare, third_class_fare]
print(fare_totals_by_class)
proportion_by_class = [fare / total_fare for fare in fare_totals_by_class]
print(proportion_by_class)

df['Age Group'] = pd.cut(df['Age'], bins=[0, 18, 64, 100], labels=['child', 'adult', 'senior'], right=False)
total_passengers = df.shape[0]
print(total_passengers)
child_count = df[df['Age Group'] == 'child'].shape[0]
print(child_count)
age_group_counts = df['Age Group'].value_counts()
print(age_group_counts)
age_group_proportions = age_group_counts / total_passengers
page_group_percentage = age_group_proportions * 100
print(page_group_percentage)







```
